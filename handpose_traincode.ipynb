```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo26n-pose.pt")  # load a pretrained model (recommended for training)

# Train the model
results = model.train(
    data="hand-keypoints.yaml",
    epochs=30,
    imgsz=640,
    device=[0, 1],    # Target both T4 GPUs (Distributed Data Parallel)
    half=True,        # Use FP16 mixed precision to speed up training and save VRAM
    workers=4,        # Use 4 dataloader workers as requested
    batch=64,         # Increased batch size to fully utilize the 30GB combined VRAM
)
```
